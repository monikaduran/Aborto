## Eliminar Datos perdidos

```python
# Existen valores NA (Valores NO asignados). Para saber cuantos hay por columna
datos.isna().sum()
```

```python
# # Existen valores NA (Valores NO asignados). Para saber porcentaje por columna
((datos.isnull().sum() / len(datos))*100)
```

```python
# Existen valores NA (Valores NO asignados). Para Eliminarlos
datos = datos.dropna(subset=["title_year"])
```
